```python
import numpy as np
import pandas as pd
```

```python
file="C:\\Git\\DataScience\\drinks.csv"
drinks = pd.read_csv(file)
drinks.columns
```

```python
drinks.head(20)
```

```python
#Apagando uma coluna temporariamente (sem usar o "inplace")

drinks.drop('continent',axis=1).head()
```

Quando referenciamos linhas e colunas com o parâmetro axis:

* axis 0 = linhas
* axis 1 = colunas

```python
#Apagando uma linha temporariamente

drinks.drop(2, axis=0).head()
```

Para apagar em definitivo, uma alternativa é armazenar o conteúdo em uma variável:

```python
A = drinks.drop('continent',axis=1).head()
```

```python
A.head()
```

## Calculando colunas numéricas

```python
#Calculando a média de cada coluna numérica

drinks.mean()
```

```python
#Repetindo o cálculo explicitando o eixo

drinks.mean(axis=0)
```

```python
drinks.describe()
```

```python
#Calculando a média para cada coluna

drinks.mean(axis=1).head()
```

```python
#Index é um alias para axis 0

drinks.mean(axis='index')
```

```python
#Columns é um alias para axis 1

drinks.mean(axis='columns').head()
```

```python
file = "C:\\Git\\DataScience\\eixos.csv"
eixos = pd.read_csv(file)
eixos.columns
```

## Trabalhando com as linhas: axis = 0

### O comportamento padrão da função mean() é o eixo 0 - axis = 0

```python
eixos.mean()
```

```python
eixos.mean(axis=0)
```

```python
eixos.sum(axis=0)
```

## Trabalhando com as colunas

```python
eixos.mean(axis = 1)
```

```python
eixos.sum(axis = 1)
```
